```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
k = 8.99e9 #electrostatic constant
eps0 = 8.85e-12 #permittivity constant
e = 1.602e-19 #electron charge
me = 9.109e-31 #electron mass
```

```python
# 22.28

q = 1.0 #test charge of 1C; result (xmax) is independent of this number! But not Emax.
R = 2.4e-2 #2.4cm in m

#probe distances from -0.15 to +0.15m in 0.001m increments:
z = np.arange(-0.15, 0.15, 0.001)
E = k*q*z/(z**2+R**2)**(3./2.)

plt.plot(z, E)
plt.grid()
plt.xlabel('z (m)')
plt.ylabel('E (N/C)')

#find maximum
print 'xmax = {0:.3f} m'.format(z[E == max(E)][0])
print 'Emax = {0:.3e} N'.format(E[E == max(E)][0])
```

```python
# 22.48

sigma = 4e-6

const =  e*sigma/(2*me*eps0)

print "const. = {0:.2e} m/s^2".format(const)

for RZ in np.array([1., 100., 1000.]):
    a = const*(1 - 1/np.sqrt(1+RZ**2))
    print "R/Z={0:4.0f}: {1:.2e} m/s^2".format(RZ, a)
```

```python
# 22.78

alpha = np.arange(0, 4, 0.001)
Ec = alpha/(1+alpha**2)**(3./2.) #E/constant

plt.plot(alpha, Ec)
plt.grid()
plt.xlabel('alpha')
plt.ylabel('E/const.')

#find alpha at max E/c
alphamax = alpha[Ec == max(Ec)][0]
print 'alphamax = {0:.3f} m'.format(alphamax)

#find alpha at halfmax
alphahalfmax1 = np.interp(0.5*max(Ec), Ec[alpha<alphamax], alpha[alpha<alphamax])
alphahalfmax2 = np.interp(-0.5*max(Ec), -1.*Ec[alpha>alphamax], alpha[alpha>alphamax])

print 'alphahalfmax1 = {0:.4f} m'.format(alphahalfmax1)
print 'alphahalfmax2 = {0:.4f} m'.format(alphahalfmax2)
```
